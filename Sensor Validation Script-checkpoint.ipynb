```python
# Importing necessary libraries for data manipulation and visualization
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Reading the uploaded CSV file into a DataFrame
file_path = 'C:\\Users\\Talha\\Downloads\\Py Practice\\Sensor Validation.csv'
df_raw = pd.read_csv(file_path)
# Displaying the first few rows of the DataFrame to get a sense of the data
df_raw.head()
```

```python
# Converting all datetime columns to standardized datetime format
datetime_columns = ['Datetime', 'Datetime1', 'Datetime3', 'Datetime4', 'Datetime 5']
```

```python
for column in datetime_columns:
    df_raw[column] = pd.to_datetime(df_raw[column], errors='coerce')
# Displaying the data types to confirm datetime conversion
df_raw.dtypes
```

```python
# Function to align data at various time intervals
def align_data(df, datetime_cols, sensor_cols, time_interval):
    # Create an empty DataFrame to store aligned data
    aligned_df = pd.DataFrame()
    
    for dt_col, sensor_col in zip(datetime_cols, sensor_cols):
        temp_df = df[[dt_col, sensor_col]].dropna()
        temp_df.set_index(dt_col, inplace=True)
        temp_df = temp_df.resample(time_interval).mean()
        aligned_df = pd.concat([aligned_df, temp_df], axis=1)
    
    return aligned_df
```

```python
# Columns containing datetime and sensor data
datetime_cols = ['Datetime', 'Datetime1', 'Datetime3', 'Datetime4', 'Datetime 5']
sensor_cols = ['PK57', 'AQM 3003', 'AQM 7003', 'Sensirion', 'BAM']
```

```python
# Align the data at 1Hr intervals
df_1Hr_aligned = align_data(df_raw, datetime_cols, sensor_cols, '1H')
```

```python
# Display the first few rows of the aligned data
df_1Hr_aligned.head()
```

```python
# Function to create time series plots for each sensor with reference to BAM
def plot_time_series(df, reference_col='BAM'):
    plt.figure(figsize=(15, 10))
    for col in df.columns:
        if col != reference_col:
            plt.subplot(2, 2, df.columns.tolist().index(col) + 1)
            plt.plot(df.index, df[col], label=col, linestyle='-', marker='o')
            plt.plot(df.index, df[reference_col], label=reference_col, linestyle='--', marker='x')
            plt.title(f'Time Series Plot: {col} vs {reference_col}')
            plt.xlabel('Datetime')
            plt.ylabel('PM2.5 (ug/m3)')
            plt.legend()
            plt.xticks(rotation=45)
            plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
# Create time series plots for 5-minute aligned data
plot_time_series(df_1Hr_aligned)
```

```python
# Function to create a single time series plot for all sensors with reference to BAM
def plot_single_time_series(df, reference_col='BAM'):
    plt.figure(figsize=(18, 8))
    
    # Generate a list of unique colors for the plot lines
    colors = ['b', 'g', 'r', 'c', 'm', 'y', 'k']
    
    # Plot the reference column first
    plt.plot(df.index, df[reference_col], label=reference_col, color=colors[0], linestyle='-', linewidth=2)
    
    # Plot the other columns
    for i, col in enumerate(df.columns):
        if col != reference_col:
            plt.plot(df.index, df[col], label=col, color=colors[i + 1], linestyle='--', marker='o')
    
    plt.title(f'Time Series Plot: Sensors vs {reference_col}')
    plt.xlabel('Datetime')
    plt.ylabel('PM2.5 (ug/m3)')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
# Create a single time series plot for 5-minute aligned data with distinct colors for each sensor and the reference
plot_single_time_series(df_1Hr_aligned)
```

```python
# Importing additional libraries for correlation matrix and heatmap
import seaborn as sns
import numpy as np
```

```python
# Function to plot correlation matrix with heatmap
def plot_correlation_matrix(df, title='Correlation Matrix'):
    correlation_matrix = df.corr()
    plt.figure(figsize=(10, 8))
    sns.set(font_scale=1)
    sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
    plt.title(title)
    plt.show()
```

```python
# Calculate and plot the correlation matrix for 5-minute aligned data
plot_correlation_matrix(df_1Hr_aligned, title='Correlation Matrix (1Hr alignment)')
```

```python
# Function to create scatter plots for each sensor with reference to BAM
def plot_scatter_plots(df, reference_col='BAM'):
    plt.figure(figsize=(15, 10))
    for col in df.columns:
        if col != reference_col:
            plt.subplot(2, 2, df.columns.tolist().index(col) + 1)
            plt.scatter(df[col], df[reference_col], label=f'{col} vs {reference_col}')
            plt.title(f'Scatter Plot: {col} vs {reference_col} (~50 ug/m3)')
            plt.xlabel(col)
            plt.ylabel(reference_col)
            plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
# Function to create scatter plots with line of best fit and R2 value, considering missing values
def plot_scatter_with_fit_line_adjusted(df, reference_col='BAM'):
    plt.figure(figsize=(15, 10))
    
    for col in df.columns:
        if col != reference_col:
            plt.subplot(2, 2, df.columns.tolist().index(col) + 1)
            
            # Data preparation considering missing values
            paired_data = df[[col, reference_col]].dropna()
            x = paired_data[col].values.reshape(-1, 1)
            y = paired_data[reference_col].values.reshape(-1, 1)
            
            # Check if enough data points are available for linear regression
            if len(x) < 2:
                plt.title(f'Insufficient data for {col} vs {reference_col}')
                plt.xlabel(col)
                plt.ylabel(reference_col)
                plt.grid(True)
                continue
            
            # Linear regression model fitting
            model = LinearRegression()
            model.fit(x, y)
            y_pred = model.predict(x)
            
            # Calculate R2 value
            r2 = r2_score(y, y_pred)
            
            # Scatter plot
            plt.scatter(x, y, label=f'{col} vs {reference_col}', alpha=0.6)
            
            # Line of best fit
            plt.plot(x, y_pred, color='red', label=f'Fit line (R2 = {r2:.2f})')
            
            plt.title(f'Scatter Plot with Fit Line: {col} vs {reference_col} (~50 ug/m3)')
            plt.xlabel(col)
            plt.ylabel(reference_col)
            plt.legend()
            plt.grid(True)
            
    plt.tight_layout()
    plt.show()
```

```python
# Create scatter plots with lines of best fit and R2 values for the data subset around 50 ug/m3, considering missing values
plot_scatter_with_fit_line_adjusted(filtered_df_1Hr)
```

```python
# Function to create scatter plots for specified concentration ranges
def plot_scatter_for_bins(df, bins, reference_col='BAM'):
    for lower, upper in bins:
        # Filter the data for the specified concentration range
        filtered_df = df[df.apply(lambda row: any(lower <= val <= upper for val in row), axis=1)]
        
        # Check if there is enough data in the specified range
        if filtered_df.empty:
            print(f"No data available in the concentration range {lower}-{upper} ug/m3.")
            continue
        
        # Create scatter plots with lines of best fit and R2 values for this filtered data
        print(f"Scatter plots for concentration range {lower}-{upper} ug/m3:")
        plot_scatter_with_fit_line_adjusted(filtered_df, reference_col)
```

```python
# Define the concentration bins
concentration_bins = [(0, 50), (50, 100), (100, 150), (150, 200)]
```

```python
# Create scatter plots for the specified concentration bins in the 1-hour aligned data
plot_scatter_for_bins(df_1hr_aligned, concentration_bins)
```

```python
# Function to create scatter plots for specified concentration ranges with labels
def plot_scatter_for_bins_labeled(df, bins, reference_col='BAM'):
    for lower, upper in bins:
        # Filter the data for the specified concentration range
        filtered_df = df[df.apply(lambda row: any(lower <= val <= upper for val in row), axis=1)]
        
        # Check if there is enough data in the specified range
        if filtered_df.empty:
            print(f"No data available in the concentration range {lower}-{upper} ug/m3.")
            continue
        
        # Create scatter plots with lines of best fit and R2 values for this filtered data
        plt.figure(figsize=(15, 5))
        plt.suptitle(f"Scatter plots for concentration range {lower}-{upper} ug/m3:", fontsize=16)
        plot_scatter_with_fit_line_adjusted(filtered_df, reference_col)
```

```python
# Create scatter plots for the specified concentration bins in the 1-hour aligned data, with labels for the ranges
plot_scatter_for_bins_labeled(df_1hr_aligned, concentration_bins)
```

```python
# Function to save scatter plots for specified concentration ranges with labels
def save_scatter_for_bins_labeled(df, bins, reference_col='BAM', file_path_prefix='/mnt/data/scatter_plot_'):
    saved_files = []
    for lower, upper in bins:
        # Filter the data for the specified concentration range
        filtered_df = df[df.apply(lambda row: any(lower <= val <= upper for val in row), axis=1)]
        
        # Check if there is enough data in the specified range
        if filtered_df.empty:
            print(f"No data available in the concentration range {lower}-{upper} ug/m3.")
            continue
        
        # Create scatter plots with lines of best fit and R2 values for this filtered data
        plt.figure(figsize=(15, 5))
        plt.suptitle(f"Scatter plots for concentration range {lower}-{upper} ug/m3:", fontsize=16)
        plot_scatter_with_fit_line_adjusted(filtered_df, reference_col)
        
        # Save the plot
        file_path = f"{file_path_prefix}{lower}-{upper}_ug_per_m3.png"
        plt.savefig(file_path)
        saved_files.append(file_path)
        
    return saved_files
```

```python
# Save scatter plots for the specified concentration bins in the 1-hour aligned data, with labels for the ranges
saved_plot_files = save_scatter_for_bins_labeled(df_1hr_aligned, concentration_bins[:3])  # Only up to 150 ug/m3
saved_plot_files
```
